# County Business Pattern 2010

```python
import pandas as pd
import numpy as np

cbp = pd.read_csv("../../external_data/cbp/cbp10co.csv")
```

```python
col = ['fipstate', 'fipscty', 'naics', 'est']
cbp_col = cbp[col]
cbp_col = cbp_col[cbp_col['naics']=='722410']
cbp_col.columns = ['STATEFP','COUNTYFP', 'naics','est']
cbp_col.head()
```

## Combine the population data

```python
popu = pd.read_csv("../../external_data/cbp/CenPop2010_Mean_CO.txt")
popu = popu[popu['STATEFP']!=72]
# popu
df = popu.merge(cbp_col, on=['STATEFP','COUNTYFP'], how='outer')
df = df[~df['COUNAME'].isnull()]
df['est'] = df['est'].fillna(0)
df['naics'] = df['naics'].fillna(722410)
df['bar_over_popu'] = df['est'] / df['POPULATION'] * 10**6

df['bar_over_popu'].describe()
df.to_csv('plot_cbp.csv')
```
